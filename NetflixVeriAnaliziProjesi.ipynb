Adım 1: Kütüphaneleri İçe Aktarma

```python
# Temel kütüphanelerimizi import ediyoruz
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Grafiklerde Türkçe karakterlerin bozulmaması için
plt.rcParams["font.family"] = "Dejavu Sans"

print("Kütüphaneler başarıyla yüklendi.")
```

Adım 2: Veriyi Yükleme ve İlk İnceleme

```python
# Netflix verisini yüklüyoruz
df = pd.read_csv("netflix_titles.csv")

# İlk 5 satırı görelim
print("Veri setinin ilk 5 satırı:")
print(df.head())

#Veri hakkında genel bilgi
print("\n Veri seti hakkında bilgi:")
print(df.info())

#Eksik değer sayıları
print("\n Eksik değerlerin dağılımı:")
print(df.isnull().sum())
```

Adım 3: Temel İstatistiksel İnceleme

```python
print("Netflix'te toplam içerik sayısı:", df.shape[0])
print("Film sayısı:", df[df['type'] == "Movie"].shape[0])
print("Dizi sayısı:", df[df['type'] == "TV Show"].shape[0])

# Ülkeler bazında en çok içerik üreten 10 ülke
print("\n En çok içerik üreten 10 ülke:")
print(df['country'].value_counts().head(10))
```

Adım 4: Görselleştirmeler

```python
#Film ve Dizi Dağılımı
plt.figure(figsize=(6,4))
sns.countplot(data=df, x="type", palette="Set2")
plt.title("Netflix'teki Film ve Dizi Dağılımı")
plt.xlabel("İçerik Türü")
plt.ylabel("Sayı")
plt.show
```

```python
#Yıllara Göre İçerik Sayısı
plt.figure(figsize=(10,5))
df['release_year'].value_counts().sort_index().plot(kind="line")
plt.title("Yıllara Göre Netflix İçerik Dağılımı")
plt.xlabel("Yıl")
plt.ylabel("İçerik Sayısı")
plt.show()
```

```python
#En Çok İçerik Üreten Ülkeler
plt.figure(figsize=(10,6))
df['country'].value_counts().head(10).plot(kind="bar", color="skyblue")
plt.title("En Çok İçerik Üreten 10 Ülke")
plt.xlabel("Ülkeler")
plt.ylabel("İçerik Sayısı")
plt.show()
```
